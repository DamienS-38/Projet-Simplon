```python
# Connexion à la base SQLite
conn = sqlite3.connect("pme.db")
cursor = conn.cursor()

# Création des tables si elles n'existent pas
cursor.execute('''
CREATE TABLE IF NOT EXISTS Produit (
    id_produit TEXT PRIMARY KEY,
    nom TEXT,
    prix REAL
)
''')

cursor.execute('''
CREATE TABLE IF NOT EXISTS Magasins (
    id_magasin TEXT PRIMARY KEY,
    nom TEXT,
    ville TEXT
)
''')

cursor.execute('''
CREATE TABLE IF NOT EXISTS Ventes (
    id_vente TEXT PRIMARY KEY,
    id_produit TEXT,
    id_magasin TEXT,
    date_vente TEXT,
    quantite INTEGER,
    FOREIGN KEY(id_produit) REFERENCES Produit(id_produit),
    FOREIGN KEY(id_magasin) REFERENCES Magasins(id_magasin)
)
''')

# Chargement des CSV
df_produits = pd.read_csv("produits.csv")
df_magasins = pd.read_csv("magasins.csv")
df_ventes = pd.read_csv("ventes.csv")

# Insertion dans Produit (ignorer les doublons avec INSERT OR IGNORE)
df_produits.to_sql('Produit', conn, if_exists='append', index=False)
df_magasins.to_sql('Magasins', conn, if_exists='append', index=False)

# Pour Ventes, on insère ligne par ligne pour éviter les doublons sur id_vente
for _, row in df_ventes.iterrows():
    cursor.execute('''
    INSERT OR IGNORE INTO Ventes (id_vente, id_produit, id_magasin, date_vente, quantite)
    VALUES (?, ?, ?, ?, ?)
    ''', (row['id_vente'], row['id_produit'], row['id_magasin'], row['date_vente'], row['quantite']))

conn.commit()
conn.close()
```
